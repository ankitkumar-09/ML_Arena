```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
train = pd.read_csv("TRAIN.csv")

print("Shape:", train.shape)
train.head()
```

```python
train.info()
```

```python
train.isnull().sum().sum()
```

```python
class_counts = train["Class"].value_counts()
class_percent = train["Class"].value_counts(normalize=True) * 100

print(class_counts)
print(class_percent)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Class", data=train)
plt.title("Class Distribution")
plt.show()
```

```python
train.describe().T
```

```python
corr_with_target = train.corr()["Class"].sort_values(ascending=False)
print(corr_with_target)
```

```python
top_features = corr_with_target[1:16]

plt.figure(figsize=(8,6))
sns.barplot(x=top_features.values, y=top_features.index)
plt.title("Top Correlated Features with Class")
plt.show()
```

```python
variances = train.drop(columns=["Class"]).var().sort_values(ascending=False)
print(variances.head(20))
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(data=train, x="F01", hue="Class", fill=True)
plt.title("F01 Density by Class")
plt.show()
```

# EDA Conclusions
- No missing values detected.
- Balanced class distribution.
- Multiple moderately correlated features.
- Visible separation in key feature distributions.
- Dataset suitable for tree-based models.
- High model accuracy is justified by strong signal. 

```python
#Distribution curves for features with highest variance
import matplotlib.pyplot as plt
import seaborn as sns

high_var_features = [
    "F31","F38","F37","F32","F33",
    "F36","F30","F35","F34","F40","F19"
]

for feat in high_var_features:
    plt.figure(figsize=(6,4))
    sns.kdeplot(data=train, x=feat, hue="Class", fill=True)
    plt.title(f"Distribution of {feat} by Class")
    plt.show()
```
